```python
#As I understand it, creates new column that provides an indication to each entry in order_day_of_the_week as to whether or not it is a busy day
result = []

for value in ord_prods_merged["order_day_of_the_week"]:
    if value == 0 or value == 1:
        result.append("Busiest days")
    elif value == 4 or value == 3:
        result.append("Least busy days")
    else:
        result.append("Regularly busy")
```

```python
result
```

```python
ord_prods_merged['busiest_days'] = result
```

```python
# client wants 3 unique values in this column, separate hours by groups of 8
# Most orders:10,11,14,15,13,12,16,9
# Average orders:17,8,18,19,20,7,21,22
# Fewest orders:23,6,0,1,5,2,4,3
```

```python
hour_of_day = []
for value in ord_prods_merged["order_hour_of_day"]:
    if value in [10, 11, 14, 15, 13, 12, 16, 9]:
        hour_of_day.append("Most Orders")
    elif value in [17, 8, 18, 19, 20, 7, 21, 22]:
        hour_of_day.append("Average Orders")
    else:
        hour_of_day.append("Fewest Orders")
```

```python
#now to add the column
ord_prods_merged['busiest_period_of_day'] = hour_of_day
```
